```python
uber_df['Date']=pd.to_datetime(uber_df['Request timestamp'],errors='coerce')
```

```python
uber_df
```

```python
uber_df['time']=uber_df['Date'].dt.hour
```

```python
uber_df.head()
```

```python
uber_df.info()
```

```python
timings=[]
for i in uber_df['time']:
        if 3<=i<7:
            timings.append('early morning')
        elif 7<=i<12:
            timings.append('morning')
        elif 12<=i<16:
            timings.append('afernoon')
        elif 16<=i<21:
            timings.append('evening')
        elif i>=21:
            timings.append('late evening')
        elif i<3:
            timings.append('late evening')
```

```python
uber_df.info()
```

```python
len(timings)
```

```python
uber_df['session']=timings
```

```python
uber_df.head()
```

```python
sns.catplot(hue='session',x='Status',data=uber_df,kind='count')
```

- from above plot in an evening  session there were  more No cars Available compared to other sessions, in morning session there were  more cancelling the cabs  compared to other session and in morning and evening session trip completed were more compared to other sessions

```python
pd.crosstab(uber_df['Pickup point'],uber_df['session'])
```

```python
supply=[]
for i in uber_df['Status']:
    if i =='Trip Completed':
        supply.append('Supplied')
    else:
        supply.append('Demand')
```

```python
uber_df['Supply']=supply
```

```python
uber_df
```

```python
uber_df['Supply'].value_counts()
```

```python
sns.catplot(x='Supply',data=uber_df,kind='count')
```

- Due to some cars are not available and some cars were cancelled so supply-demand gap exists 
- from above plot demand of cabs  was more compared to supplied of cabs
